## Modificaci√≥n del Dataframe

#### A√±adir una columna basandose en otras

```python
df["prueba"] = df["Account length"] * 10
df["prueba"] = df["Account length"].apply(lambda x: x + 1)
df["prueba2"] = df["International plan"].map(dict(Yes=1, No=0))
```

#### Borrar variables

```python
del df['prueba'] # opci√≥n 1
```

```python
df = df.drop(['prueba2'], axis=1) # Opci√≥n 2
```

#### Cambiar tipo de la variable

```python
df['Area code'] = df['Area code'].astype(object)
```
