# <center> Visualización con Pandas
Pandas teine su propio [módulo para visualización](https://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html)

## Análisis de varaibles categóricas:  `.bar()`, `.hbar()`, `.pie()`

```python
list(df.select_dtypes(include=["category" , "object", "bool"]).columns)
list(df.select_dtypes(exclude=["category" , "object", "bool"]).columns)
```

```python
df.select_dtypes(include=['object', 'bool']).nunique()
```

```python
df["State"].value_counts().plot.bar(figsize=(10, 4));
```

```python
df["State"].value_counts().plot.barh(figsize=(10, 10));
```

```python
df["International plan"].value_counts().plot.pie();
```

## Análisis de varaibles numéricas: `.hist()`, `.kde()` y `.box()`

```python
df["Account length"].plot.hist(bins=25, figsize=(10, 4));
```

```python
df["Account length"].plot.kde(figsize=(10, 4));
```

```python
df["Account length"].plot.box(vert=False, figsize=(10, 4));
```

# <center> Análisis bivariante

## Variables Numérica + Numérica: `.scatter()` `.hexbin()`

```python
df.plot.scatter(x='Total day minutes', y='Total day calls');
```

```python
df.plot.hexbin(x='Total day minutes', y='Total day calls', gridsize=25);
```

```python
df.plot.scatter(x='Total day minutes', y='Total day charge');
```

## Variables Numérica + Categórica: `groupby` + `.hist()`

```python
df.groupby('International plan')['Account length'].plot.hist(bins=25);
```

```python
df.groupby('International plan')['Account length'].hist(bins=25, alpha=0.5, density=True);
```

```python
df.groupby('International plan')['Account length'].plot.kde();
```

## Variables: Categórica + Categórica

```python
g = df.groupby('Churn')['Customer service calls'].value_counts(normalize=True).unstack().transpose().plot.bar()
```
