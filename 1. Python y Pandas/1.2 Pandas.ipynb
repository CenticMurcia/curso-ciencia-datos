## Import Pandas

```python
import pandas as pd

pd.set_option("display.precision", 2)      # Display 2 decimals
pd.set_option('display.max_columns', 1000) # Display all variables
```

## Create dataframe `df`

```python
df = pd.read_csv('../data/telecom_churn.csv')
df.head()
```

> # Remember
> ### Column = Variable = Feature = Dimension
> ### Row = Case = Point

## See shape

```python
print("Shape:", df.shape)
print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])
```

## See variables

```python
print(df.columns)
```

## Show variable types with `info()`

```python
print(df.info())
```

## Show statistics of numeric variables

```python
df.describe()
```

## Show statistics of categoric variables

```python
df.describe(include=['object', 'bool'])
```

## Select 1 variable (serie)

```python
df.State
```

```python
df["State"]
```

## Select multiple variables

```python
df[["State", "Account length", "Area code"]]
```

## Numerical variable -> Min, max, mean, median, std, ...

```python
df['Account length'].min()
```

```python
df['Account length'].max()
```

```python
df['Account length'].mean()
```

```python
df['Account length'].median()
```

```python
df['Account length'].std()
```

## Categorical variable -> Count unique values

```python
df['Churn'].value_counts()
```

```python
df['Churn'].value_counts(normalize=True)
```

```python
# Show top 10 in a dataframe
pd.DataFrame({  'Recuento':   df['State'].value_counts(),
                'Porcentaje': df['State'].value_counts(normalize=True) * 100
             }).head(10)
```

## Categorial varaibles -> Cross table
#### Also known as Contingency table

```python
pd.crosstab(df['Churn'], df['International plan'])
```

```python
pd.crosstab(df['Churn'], df['International plan'], margins=True)
```

```python
pd.crosstab(df['Churn'], df['Voice mail plan'], normalize=True)
```

## Numeric & Categorial varaibles -> Pivot table

```python
df.pivot_table(['Total day calls', 'Total eve calls', 'Total night calls'],
               ['State'], aggfunc='mean').head()
```

```python
df.pivot_table(['Total day calls', 'Total eve calls', 'Total night calls'],
               ['State', "Churn"], aggfunc='mean').head()
```

# <center> Filters

## 1 Filter

```python
df[df['Churn'] == True]
```

## Multiple filters

```python
df[ (df['Churn']==False) & (df['International plan']=='No')]
```

# <center> Dataframe Modification

## Change type to categoric

```python
df['Area code'] = df['Area code'].astype(object)
```

# <center> Summary
- The data table is a Pandas **dataframe**
- The data column is a Pandas **series**
- We can obtain information of the **dataset**:
  - 10 top rows: `df.head(10)`
  - Number of rows & variables: `df.shape`
  - Names of variables: `df.columns`
  - Types of variables: `df.info()`
  - Stats of numerical variables: `df.describe()`
  - Stats of numerical variables: `df.describe(include=['object', 'bool'])`
- Information of **numerical variables**:
  - Minimum `df.num_var.min()`
  - Maximun `df.num_var.max()`
  - Mean `df.num_var.mean()`
  - Median `df.num_var.median()`
  - Std `df.variable.std()`
- Information of **categorial variables**:
  - Unique values (count): `df.cat_var.value_counts()`
  - Unique values (percen): `df.cat_var.value_counts(normalize=True)`
  - 2 categorial vars (count): `pd.crosstab(df.cat_var1, df.cat_var2, margins=True)`
  - 2 categorial vars (percen): `pd.crosstab(df.cat_var1, df.cat_var2, margins=True, normalize=True)`
- Information of **numerical & categorial variables**:
  - Pivot table: `df.pivot_table([num_var1, num_var2, ...], [cat_var1, cat_var2], aggfunc='mean')`
- **Fitering rows**:
  - One filter: `df[ condition ]`
  - Multiple filters: `df[ (condition1) & (condition2)]`
