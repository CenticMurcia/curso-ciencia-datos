### Create time data

```python
dates = pd.DataFrame(data={'date':pd.date_range(start='2013-01-01', end='2015-11-30')})
```

### Create feats

```python
dates['month']          = dates.date.dt.month 
dates['year']           = dates.date.dt.year - 2013
dates["day_of_week"]    = dates.date.dt.dayofweek + 1
dates["day_of_month"]   = dates.date.dt.day
dates['day_of_year']    = dates.date.dt.dayofyear
dates["day_of_history"] = dates.day_of_year + (365 * dates.year)
dates['date_block_num'] = dates['year']*12 + dates['month'] - 1

dates.set_index('date', inplace=True)

dates
```

```python
dates.agg(["min", "max"]).T
```

### Aggregate by date_block_num (month from 0 to 34)

```python
days_of_week = ["#mon", "#tue", "#wed", "#thu", "#fri", "#sat", "#sun"]

months = dates.groupby(["date_block_num", "day_of_week"]).size().unstack()
months.columns = days_of_week
months["#sat+sun"] = months["#sat"] + months["#sun"]

months["#days"] = months[days_of_week].sum(1)
months["month_id"]   = months.index%12 + 1
months
```

### Save

```python
months.to_csv(DATA_FE_PATH + "MONTHS.csv")
```
